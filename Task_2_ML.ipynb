```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
df = pd.read_csv("Dataset .csv")
```

```python
columns_to_drop = [
    'Restaurant ID', 'Restaurant Name', 'Address', 'Locality', 'Locality Verbose',
    'Switch to order menu', 'Rating color', 'Rating text'
]
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.dropna(subset=['Cuisines'], inplace=True)
```

```python
df['Combined_Features'] = (
    df['Cuisines'].astype(str) + ' ' +
    df['Price range'].astype(str) + ' ' +
    df['Has Table booking'].astype(str) + ' ' +
    df['Has Online delivery'].astype(str)
)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Combined_Features'])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
df = df.reset_index()
```

```python
def recommend_restaurants(user_input, top_n=5):
    # Convert input into the same format as Combined_Features
    user_profile = (
        user_input['Cuisines'] + ' ' +
        str(user_input['Price range']) + ' ' +
        str(user_input['Has Table booking']) + ' ' +
        str(user_input['Has Online delivery'])
    )

    user_vec = tfidf.transform([user_profile])
    similarity_scores = cosine_similarity(user_vec, tfidf_matrix).flatten()
    top_indices = similarity_scores.argsort()[-top_n:][::-1]
    print("\nRecommended Restaurants Based on Your Preferences:\n")

    # Move the loop inside the function
    for idx in top_indices:
        print(f"Restaurant {idx + 1}:")
        print(f"  Cuisines: {df.loc[idx, 'Cuisines']}")
        print(f"  Price Range: {df.loc[idx, 'Price range']}")
        print(f"  Table Booking: {df.loc[idx, 'Has Table booking']}")
        print(f"  Online Delivery: {df.loc[idx, 'Has Online delivery']}")
        print(f"  Rating: {df.loc[idx, 'Aggregate rating']}")
        print("-" * 50)
```

```python
sample_input = {
    'Cuisines': 'North Indian',
    'Price range': 3,
    'Has Table booking': 1,
    'Has Online delivery': 1
}
```

```python
recommend_restaurants(sample_input)
```
